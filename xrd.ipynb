```python
from pymatgen.ext.matproj import MPRester 
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt

with MPRester("LUZDaZywJ3wcPUgp0h4") as m:

    # Structure for material id
    struct = m.get_structure_by_material_id("mp-1023936")
```

# Tâche 7
Dans ce notebook nous allons déterminer les 3 premiers pics du diffractogramme correspondant au matériau WSe2  pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å. Grâce aux phénomènes d'interférences (constructives ou destructives en fonction des directions), nous pouvons obtenir une image de diffraction qui, sous certaines hypothèses, peut nous permettre de déterminer la position des atomes de la structure.

```python
xrd = XRDCalculator(1.54060, symprec = 0, debye_waller_factors=None)
pattern = xrd.get_pattern(struct)
```

## Premier Pic


```python
#angle premier pic
theta_1 = pattern.x[0]
#intensité
Int1 = pattern.y[0]
#indices hkl
miller1 = pattern.hkls[0]

print("Angle (2\u0398) premier pic : {}°".format(theta_1))
print("avec pour indices hkl :" + str(miller1))
print("et une intesité du pic de " + str(Int1))
```

## Deuxième Pic

```python
#angle second pic
theta_2 = pattern.x[1]
#intensité
Int2 = pattern.y[1]
#indices hkl
miller2 = pattern.hkls[1]

print("Angle (2\u0398) deuxième pic : {}°".format(theta_2))
print("avec pour indices hkl :" + str(miller2))
print("et une intesité du pic de " + str(Int2))
```

## Troisième Pic

```python
#angle troisième pic
theta_3 = pattern.x[2]
#intensité
Int3 = pattern.y[2]
#indices hkl
miller3 = pattern.hkls[2]

print("Angle (2\u0398) troisième pic : {}°".format(theta_3))
print("avec pour indices hkl :" + str(miller3))
print("et une intesité du pic de " + str(Int3))
```

## Affichage des trois premiers pics sur le diffractogramme
Pour finir nous pouvons afficher le diffractogramme de notre matériau qui représente l'angle 2xtheta en fonction de l'intensité (mise à l'échelle de 0 à 100).

```python
xrd.show_plot(struct,two_theta_range=(0,20),annotate_peaks=True,with_labels=True)
```
